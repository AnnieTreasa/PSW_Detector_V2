```python
# Creating Raw Matrices of Audio Files - NORMALISATION
norm_array = []
def create_fold_spectrograms(audio_path):
    for audio_file in list(Path(audio_path).glob('*.wav')):

            samples, sample_rate = librosa.load(audio_file)
            S = librosa.feature.melspectrogram(y=samples, sr=sample_rate)
            feature = S

            #print("S : ",S)
            max_value = np.amax(S)
            #print("Max of S :",max_value)
            min_value = np.amin(S)
            #print("Min of S :",min_value)

            max_val=float("-INF")
            for i in feature:
              max_val=max(max(i),max_val)
            f=feature[:]
            for i in range(len(f)):
              for j in range(len(f[0])):
                if f[i][j]<0:
                  f[i][j]=0
                else:
                  f[i][j]=f[i][j]/max_val*255
            norm_array.append(f)
            #print("f : ",f)
            max_value = np.amax(f)
            #print("Max of f :",max_value)
            min_value = np.amin(f)
            #print("Min of f :",min_value)
            i=i+1
    return norm_array

audio_path = Path('/content/drive/My Drive/Dataset/mixdata_100')
norm_array=create_fold_spectrograms(audio_path)

n=len(norm_array)
print(n)
```

```python
import numpy as np

def cosine_similarity(x, y):

    # Ensure length of x and y are the same
    if len(x) != len(y) :
        return None

    # Compute the dot product between x and y
    x = x.astype(np.ulonglong)
    y = y.astype(np.ulonglong)
    dot_product = np.sum(x*y)

    # Compute the L2 norms (magnitudes) of x and y
    magnitude_x = np.sum(x*x)
    magnitude_x = np.sqrt(magnitude_x)

    magnitude_y = np.sum(y*y)
    magnitude_y = np.sqrt(magnitude_y)

    # Compute the cosine similarity
    cosine_similarity = dot_product / (magnitude_x * magnitude_y)

    return cosine_similarity
```

```python
import os
import shutil

def copy_files(source_folder, destination_folder, num_files):
    # Get a list of files in the source folder
    file_list = os.listdir(source_folder)


    # Iterate over the first num_files files
    for filename in file_list[:num_files]:
        source_file = os.path.join(source_folder, filename)
        destination_file = os.path.join(destination_folder, filename)

        # Copy the file to the destination folder
        shutil.copy(source_file, destination_file)

source_folder = "/content/drive/My Drive/Dataset/spectro_data/nopsw_spectro"
dest_folder = "/content/drive/My Drive/Dataset/MIXDATA_100_spectro"
copy_files(source_folder,dest_folder,50)
```

```python
import numpy as np

def cc(x, y):

    # Ensure length of x and y are the same
    if len(x) != len(y) :
        return None

    # Compute the dot product between x and y
    x = x.astype(np.ulonglong)
    y = y.astype(np.ulonglong)
    xm=np.mean(x)
    ym=np.mean(y)
    dot_product = np.sum((x-xm)*(y-ym))

    # Compute the L2 norms (magnitudes) of x and y
    magnitude_x = np.sum((x-xm)*(x-xm))
    magnitude_x = np.sqrt(magnitude_x)

    magnitude_y = np.sum((y-ym)*(y-ym))
    magnitude_y = np.sqrt(magnitude_y)

    # Compute the cosine similarity
    co_rel = dot_product / (magnitude_x * magnitude_y)

    return co_rel
```
